**Animals**

```python
#Create a common animals class
class animal:
    def __init__(self,legs,sound,wings=False):
        self.legs = legs
        self.sound = sound
        self.wings = wings
        #declare a protected variable
        
    def get_legs(self):
        print("{} has {} legs.".format(self.__str__(), self.legs))
    
    def speak(self):
        print("{}".format(self.sound))
        
    def print_unique(self):
        print(self.get_unique())
    
class bat(animal):
    def __init__(self, legs=2, sound='screech', wings=True):
        animal.__init__(self, legs, sound, wings)
        #every animal has a protected 'unique' variable
        self.__unique = "Can't see. Uses ultrasound to search for prey"
        
    def __str__(self):
        return "Bat"
    
    def get_unique(self):
        return self.__unique
    
class bear(animal):
    def __init__(self, legs=4, sound='growl', wings=False):
        animal.__init__(self, legs, sound, wings)
        #every animal has a protected 'unique' variable
        self.__unique = "Love to eat Fish!"
        
    def __str__(self):
        return "Bear"
    
    def get_unique(self):
        return self.__unique
    
class cat(animal):
    def __init__(self, legs=4, sound='meow', wings=False):
        animal.__init__(self, legs, sound, wings)
        #every animal has a protected 'unique' variable
        self.__unique = "I can also purr!"
        
    def __str__(self):
        return "Cat"
    
    def get_unique(self):
        return self.__unique
    
class cow(animal):
    def __init__(self, legs=4, sound='moo', wings=False):
        animal.__init__(self, legs, sound, wings)
        #every animal has a protected 'unique' variable
        self.__unique = "Gives you Milk!"
        
    def __str__(self):
        return "Cow"
    
    def get_unique(self):
        return self.__unique
    
class dog(animal):
    def __init__(self, legs=4, sound='bark', wings=False):
        animal.__init__(self, legs, sound, wings)
        #every animal has a protected 'unique' variable
        self.__unique = "Man's best friend"
        
    def __str__(self):
        return "Dog"
    
    def get_unique(self):
        return self.__unique
    
class dove(animal):
    def __init__(self, legs=2, sound='coo', wings=True):
        animal.__init__(self, legs, sound, wings)
        #every animal has a protected 'unique' variable
        self.__unique = "Represent peace!"
        
    def __str__(self):
        return "Dove"
    
    def get_unique(self):
        return self.__unique

class elephant(animal):
    def __init__(self, legs=4, sound='trumpet', wings=False):
        animal.__init__(self, legs, sound, wings)
        #every animal has a protected 'unique' variable
        self.__unique = "Big ears!"
        
    def __str__(self):
        return "Elephant"
    
    def get_unique(self):
        return self.__unique
    
class frog(animal):
    def __init__(self, legs=4, sound='croak', wings=False):
        animal.__init__(self, legs, sound, wings)
        #every animal has a protected 'unique' variable
        self.__unique = "Loves swampy places!"
        
    def __str__(self):
        return "Frog"
    
    def get_unique(self):
        return self.__unique
    
class grasshopper(animal):
    def __init__(self, legs=6, sound='chirp', wings=True):
        animal.__init__(self, legs, sound, wings)
        #every animal has a protected 'unique' variable
        self.__unique = "Greeeeeen!"
        
    def __str__(self):
        return "Grasshopper"
    
    def get_unique(self):
        return self.__unique
    
class owl(animal):
    def __init__(self, legs=2, sound='hoot', wings=True):
        animal.__init__(self, legs, sound, wings)
        #every animal has a protected 'unique' variable
        self.__unique = "Big eyes, can see it in the night!"
        
    def __str__(self):
        return "Owl"
    
    def get_unique(self):
        return self.__unique
        
        
```

**Examples of abstraction**

```python
# Instances of following animals are generated.
b1 = bat()
c1 = cat()
d1 = dog()
# We can access public variables
print((c1.legs, c1.wings))
# We can call methods of parent class which are abstracted in 'animals' class.
b1.get_legs()
d1.speak()
c1.print_unique()
```

**Shapes**

```python
#Base class
class Shapes:
    def __init__(self):
        self._Type = '' #Protected member
        print("A shape has been created!")

    def get_Type(self):
        print(self._Type)
        
    def __del__(self):
        print('Shape has been destructed!')
        
#Triangle
class triangle(Shapes):
    def __init__(self):
        self._side = 3
    #We can modify protected members via base class    
    def set_type(self):
        self._Type = 'Polygon with 3 sides'
    #Only member functions can have access to private members    
    def get_side(self):
        print(self._side)
        
#Equilateral traingle inherited from triangle
class equilateral(triangle):
    #We can modify protected members via base class    
    def set_type(self):
        self._Type = 'Polygon with 3 sides and all three are equal'
    #Only member functions can have access to private members    
    def get_side(self):
        print(self._side)
        
#Rectangle
class rectangle(Shapes):
    def __init__(self):
        self.__side = 4
    #We can modify protected members via base class    
    def set_type(self,typ):
        self._Type = 'Polygon with 2 equal and 2 unequal sides'
    #Only member functions can have access to private members    
    def get_side(self):
        print(self.__side)
        
#Square inherited from rectangle
class square(rectangle):
    def __init__(self):
        self.sides = 4
    #We can modify protected members via base class    
    def set_type(self):
        self._Type = 'Polygon with 4 sides all of which are equal'
    
    def get_side(self):
    #The getter function will throw an error if this is executed as the private member of base class is being accessed
    #print(self.__side)
        print(self.sides)
        
#Rhombus inherited from square
class rhombus(square):
    def set_type(self):
        self._Type = 'Polygon with 4 sides all of which are equal but angles are not 90'
    
    #The base class has a public member in sides so we can access it in the base class and outside
    def get_side(self):
        print(self.sides)
        
#Circle
class circle(Shapes):
    def __init__(self):
        Shapes.__init__(self)
        self.__r = 0
        self.__eqn = 'x^2 + y^2 = r^2'
    def set_type(self):
        self._Type = 'Conic'
    #Setter function for accessing private member
    def set_radius(self,r):
        self.__r = r
    def get_area(self):
        print(3.14 * (self.__r)**2)
        
#Ellipse
class ellipse(Shapes):
    def __init__(self):
        self.__a = 0
        self.__b = 0
        self.__eqn = 'x^2/a^2 + y^2/b^2 = 1'
    def set_type(self):
        self._Type = 'Conic with a major and minor axis'
    #Setter function for accessing private member
    def set_radius(self,a,b):
        self.__a = a
        self.__b = b
    def get_area(self):
        print(3.14 * self.__a * self__b)
```

**Example of destructor usage**

```python
#Constructor called while creating an instance of child class "Circle"
c1 = circle()

#Destructor called while deleting an instance of child class
del c1

#Both the methods have been inherited from parent class Shapes.
```
